## Selecting a single column

Simplest case: selecting a single column by column name.

General syntax:
```
df['column_name']

This is an example of **label-based subsetting**, which means we want to select data from our data frame using the **names** of the columns, **not their position**.

````python
# Example 

# Select number of bird species observed at Mugu Lagoon in spring
mul_spring = df['MUL_spring']

mul_spring #this is panda series
````

````python
type(mul_spring)
````

We can do label-based subsetting of a single column using attribute syntax

`df.column_name`

### Example

````python
df.MUL_spring
````

## Selecting multiple columns...

... using a list of column names.

Syntax:

```
df[ ['column_1', 'column_10', 'column_245'] ]

Notice there are double square brackets. This is because we are passing the list of names ['column_1', 'column_10', 'column_245'] to the selection brackets [].

### Example

````python
# Select species abundance in Tijuana Estuary during winter and fall
tje_wf = df[['TJE_winter','TJE_fall']]
````

````python
# Check the data type of tje_wf

type(tje_wf)
````

````python
# Check the shape of the selection

tje_wf.shape
````

## .. using a slice

To select a slice of the columns we will use a special case of `loc` selection.

Syntax:
```
df.loc[ : , 'column_start': 'column_end']

'column_start': 'column_end' = starting and ending point of column slice, slice includes both endpoints.

- the first value passed to `loc` is used to select rows, using a `:` as the row-selection parameter means "select all the rows"

### Example

````python
# Selcet columns between CSM_winter and MUL_fall
csm_mul = df.loc[:,'CSM_winter':'MUL_fall']
csm_mul.head()
````

### Selecting rows ...

### ... using a condition

Syntax
```
df[condition_on_rows]
```

That `condition_on_rows` can be many things.

### Example

We are intrested in all data after 2020

````python
# Select all rows with year > 2020
post_2020 = df[df['year'] > 2020]
post_2020
````

condition for ou rows = `df['year']>2020` this is a `panda.Series` with **boolean values** (True or False) indicates which row satisfy the condition year>2020

````python
# Check the type of df['year']>2020
print(type(df['year']>2020))

# Print the boolean series
df['year']>2020     
````

````python
### Check in
#Get the subset of the dataframe on which San Dieguito Wetland has atleast 75 species recorded during spring.

sw_subset = df[df['SDW_spring'] >= 75]
sw_subset
````

### Example

Supose we want to look at the data from 2012 to 2015 (including both years).

````python
subset = df[df['year'].between(2012, 2015)]
subset
````

 - `df['year'] = column with year values, this is a panda.Series

- in df['year'].between(), we have that `between()` is a method for the pandas.Series and we are calling it using the dot .

- (2012, 2015) are the parameters for the between() method, from the pandas documentation we can see this method will subset including both endpoints

- df['year'].between(2012, 2015) is then a pandas.Series of boolean values indicating which rows have year equal to 2012, 2013, 2014, or 2015.

- when we put df['year'].between(2012, 2015) inside the selection brackets [] we obtain the rows of the data frame with year equal to 2012, 2013, 2014, 2015.

## Avoid using 'loc` for selecting only rows

It is equivalent to write

```
Select rows with year < 2015
df[df['year']<2015]
```

and

```
df.loc[df['year'],2015, :]
```

In the second one:
- using `df.loc[ row selection, column-selection]`
- the row selection is the condition `df['year']<2015`
- the column selection is `:` = all columns


## ... using multiple conditions

We can combine conditions to select rows by surrounding each in parenthesis `()` and using the or operator `|` and the operator '&'.

### Example: or

````python
#Select rows in CSM with more than 50 species in winter or fall
df[ (df['CSM_winter']>50) | (df['CSM_fall']>50)]
````

### Select rows in which both the CSM and the SDW have more than 60 bird species during spring

````python
df[(df['CSM_spring']>60) & (df['SDW_spring']>60)]
````
